```python
profiles = [
    'https://x.com/realdonaldtrump?lang=en',
    'https://x.com/JoeBiden',
    'https://x.com/imVkohli',
    'https://x.com/narendramodi',
    'https://x.com/RahulGandhi',
    'https://x.com/KicchaSudeep'
]

threads = []
```

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.actions.wheel_input import ScrollOrigin
from selenium.webdriver.support.relative_locator import locate_with

import time
import threading


driver = webdriver.Chrome()
driver.get("http://127.0.0.1:8000")
driver.minimize_window()
driver.maximize_window()
```

```python
# Open multiple windows
for i in range(4):
    driver.switch_to.new_window('tab')
    driver.get("http://127.0.0.1:8000")
    time.sleep(0.4)
```

```python
# Start entering urls for each window
start, stop = 0, 0
timers = []
for i, window in enumerate(driver.window_handles):
    driver.switch_to.window(window)
    urlEle = driver.find_element(By.ID, "search")
    urlEle.send_keys(profiles[i])
    start = time.time()
    urlEle.send_keys("\uE007")
    stop = time.time()
    timers.append(stop - start)
    time.sleep(1)
```

```python
print("\t\tTIME TAKEN FOR EACH TEST CASE\n")
print("\tTEST CASE\t\tTime Taken (in seconds)")

for i, timetaken in enumerate(timers):
    print("\t", i, "\t\t\t", timetaken)
```

```python
import matplotlib.pyplot as plt
import numpy as np

accuracy = np.array([timers[0]])
labels = ["100%", "75%", "50%"]
```

```python
plt.pie(accuracy, labels = [labels[0]], colors = ["green"])
plt.title("ACCURACY OF THE INPUTS vs OUTPUTS")
plt.show() 
```

```python
timerpoints = np.array(timers)

plt.plot(timerpoints, color = 'r')
plt.xlabel("Test Cases")
plt.ylabel("Time in seconds")
plt.title("OVERALL TIME TAKEN FOR EACH TEST CASE")
plt.show()
```
